# Exercise 1

* Part 1 - Implement the [21 game](https://en.wikipedia.org/wiki/Nim#The_21_game), so that you can play with the computer. 
    * The play should be interactive, and the winner should be announced.
    * Make sure not to go to negative numbers.
* Part 2 - Can you implement it, so that if the computer is playing second, it will always wins?

```
from random import choice

choice(range(13))
```

## Solution

```
n = 21
min_take = 1
max_take = 3
```

### Part 1

```
from random import choice
```

```
while n:
    print(n)
    my_turn = int(input("How many coins do you want to take?\n"))
    while my_turn not in range(min_take, min(max_take, n)+1):
        my_turn = int(input(f"You must take between {min_take} and {min(max_take, n)}. Please choose again."))
    n -= my_turn
    if n==0:
        print("You win!")
        break

    print(n)
    pc_turn = choice(range(min_take, min(max_take, n)+1))
    print(f"I take {pc_turn}")
    n -= pc_turn
    if n==0:
        print("PC wins!")
        break
```

# Exercise 2

Solve the following exercises of [verbal arithmatics](https://en.wikipedia.org/wiki/Verbal_arithmetic):
* $4 \times ABCD=DCBA$
* $4 \times ABCDE = EDCBA$
* $SEND + MORE = MONEY$

## Solution

```
for ABCD in range(1234, 9876):
    DCBA = int(str(ABCD)[::-1])
    if 4 * ABCD == DCBA:
        print(f"4 x {ABCD} = {DCBA}")
```

```
for ABCDE in range(12345, 98765):
    EDCBA = int(str(ABCDE)[::-1])
    if 4 * ABCDE == EDCBA:
        print(f"4 x {ABCDE} = {EDCBA}")
```

```
for SEND in range(1000, 9876):
    # if SEND%1000==0:
    #     print(f"SEND={SEND}")
    E = SEND // 100 % 10
    N = SEND // 10 % 10
    for MOR in range(100, 987):
        M = MOR // 100
        O = MOR // 10 % 10
        R = MOR % 10
        MORE = 10*MOR + E
        for Y in range(10):
            MONEY = 10000*M + 1000*O + 100*N + 10*E + Y
            if SEND + MORE == MONEY:
                digits = set(str(SEND)+str(MORE)+str(MONEY))
                if len(digits)==8:
                    print(f"{SEND}+{MORE}={MONEY}")
```

# Exercise 3

Find a number which is both a [triangular number](https://en.wikipedia.org/wiki/Triangular_number) and a square number.

## Solution

```
N = 10000
squares = set([x**2 for x in range(N)])
triangles = set([sum(range(x)) for x in range(N)])
print(sorted(squares & triangles))
```
